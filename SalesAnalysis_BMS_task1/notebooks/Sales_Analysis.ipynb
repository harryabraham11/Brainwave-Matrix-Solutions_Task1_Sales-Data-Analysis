An Exploratory Sales Data Analysis for a Supermarket

```python
#importing pandas library
import pandas as pd 
```

```python
#loading the dataset
data_frame = pd.read_csv("../data/Sample - Superstore.csv", encoding='latin1')
```

```python
#preview of the Data
data_frame.head()
```

```python
#Obtaining the basic Info of the Dataset
data_frame.info()
```

```python
#Summary statistics for numerical columns
data_frame.describe(include='all')
```

```python
# Checking for null values column-wise
data_frame.isnull().sum()
```

```python
#Checking for duplicate rows
data_frame.duplicated().sum()
data_frame.drop_duplicates(inplace=True)
```

```python
# Converting the  date columns to datetime format
data_frame['Order Date'] = pd.to_datetime(data_frame['Order Date'])
data_frame['Ship Date'] = pd.to_datetime(data_frame['Ship Date'])
```

```python
duplicates = data_frame.duplicated()
print("Number of duplicate rows:", duplicates.sum())
```

```python
#Order Month
data_frame['Order Month'] = data_frame['Order Date'].dt.to_period('M')
```

```python
#Shipping duration
data_frame['Shipping Duration'] = (data_frame['Ship Date'] - data_frame['Order Date']).dt.days
```

```python
#Profit Margin
data_frame['Profit Margin'] = (data_frame['Profit'] / data_frame['Sales']).round(2)
```

```python
#Sales per Unit
data_frame['Sales per Unit'] = (data_frame['Sales'] / data_frame['Quantity']).round(2)
```

```python
#Discount Flag
data_frame['Is Discounted'] = data_frame['Discount'].apply(lambda x: 1 if x > 0 else 0)
```

```python
#importing Matplotlib and Seaborn libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Making the plots look better
plt.style.use('ggplot')
sns.set_palette('Set2')
```

```python
#Sales and Profit Overview
print("Total Sales: â‚¹", round(data_frame['Sales'].sum(), 2))
print("Total Profit: â‚¹", round(data_frame['Profit'].sum(), 2))
print("Total Quantity Sold:", data_frame['Quantity'].sum())
print("Average Discount:", round(data_frame['Discount'].mean(), 2))
```

```python
#Sales and profit by category
category_group = data_frame.groupby('Category')[['Sales', 'Profit']].sum().sort_values('Sales', ascending=False)
category_group.plot(kind='bar', figsize=(8,5), title='Sales vs Profit by Category')
plt.ylabel('Amount in Rs')
plt.savefig('Reports/sales_by_category.png')
plt.show()
```

```python
#Top 10 Products by Sales
top_products = data_frame.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='barh', figsize=(8,5), title='Top 10 Products by Sales')
plt.xlabel('Sales')
plt.show()
```

```python
#Monthly Sales Trend
monthly_sales = data_frame.groupby('Order Month')['Sales'].sum()
monthly_sales.plot(kind='line', marker='o', figsize=(10,5), title='Monthly Sales Trend')
plt.ylabel('Sales')
plt.xticks(rotation=0)
plt.show()
```

```python
#Segment-wise Profitability
segment_profit = data_frame.groupby('Segment')['Profit'].sum()
segment_profit.plot(kind='pie', autopct='%1.1f%%', startangle=90, figsize=(6,6), title='Profit Distribution by Segment')
plt.ylabel('')
plt.show()
```
